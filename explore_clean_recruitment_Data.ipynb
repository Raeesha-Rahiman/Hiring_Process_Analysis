From above observation 
- Multiple rows per vacancy capture different lifecycle events, but are fragmented.
- Some rows have Approved and Sourcing start dates, others have Interview start or Filled.
- Status column is inconsistent across duplicates for the same vacancy.

```python
#Since each row holds partial info, will  merge duplicate rows into one, picking the right date for each event.
agg_dict = {
    'Employment_Type': 'first',
    'BU Region': 'first',                   #Approved, Sourcing start — take the earliest (minimum) date, since those happen once at start.
    'Approved': 'min',                      #Interview end, Offered, Filled — take the latest (maximum) date, as they happen last
    'On hold': 'min',                       #Employment_Type, BU Region — pick the first non-null value (usually all same).
    'Sourcing start': 'min',
    'Interview start': 'min',
    'Interview end': 'max',
    'Offered': 'max',
    'Filled': 'max'
}

df_cleaned = df.groupby('ID').agg(agg_dict).reset_index()
```

```python
def assign_status(row):
    if pd.notnull(row['Filled']):
        return 'Filled'
    elif pd.notnull(row['Offered']):
        return 'Offered'
    elif pd.notnull(row['Interview end']):
        return 'Interview end'
    elif pd.notnull(row['Interview start']):
        return 'Interview start'
    elif pd.notnull(row['Sourcing start']):
        return 'Sourcing start'
    elif pd.notnull(row['On hold']):
        return 'On hold'
    elif pd.notnull(row['Approved']):
        return 'Approved'
    else:
        return 'Unknown'

df['Status'] = df.apply(assign_status, axis=1)
```

```python
df['Approved <= On hold'] = (df['Approved'] <= df['On hold']) | (df['On hold'].isna())
df['On hold <= Sourcing start'] = (df['On hold'] <= df['Sourcing start']) | (df['Sourcing start'].isna())

print(df[['Approved <= On hold', 'On hold <= Sourcing start']].all())
```

```python
df_cleaned['Interview start <= Interview end'] = (df_cleaned['Interview start'] <= df_cleaned['Interview end']) | (df_cleaned['Interview end'].isna())
df_cleaned['Offered <= filled'] = (df_cleaned['Offered'] <= df_cleaned['Filled']) | (df_cleaned['Filled'].isna())

print(df_cleaned[['Interview start <= Interview end', 'Offered <= filled']].all())
```

```python
df_cleaned['Valid_Offer_Fill'] = (df['Offered'] <= df_cleaned['Filled']) | (df_cleaned['Filled'].isna()) | (df_cleaned['Offered'].isna())
print(df_cleaned['Valid_Offer_Fill'].all())  # Should be True if data is consistent
```
