```python
conn = engine.connect()
```

```python
def ingest_db(df, table_name, engine):
    df.to_sql(table_name, con=engine, if_exists='append', index=False)

for file in os.listdir('Hiringdata'):
    if file.endswith('.csv'):
        print(f"Processing file: {file}")
        table_name = file[:-4]
        df = pd.read_csv(f'Hiringdata/{file}')
        ingest_db(df, table_name, engine)
```

```python
table_list = tables_df.iloc[:, 0].tolist()

table_count = {}

# Loop through each table and get count
for table in table_list:
    count_df = pd.read_sql(f"SELECT COUNT(*) AS count FROM {table}", conn)
    table_count[table] = count_df.iloc[0, 0]  # ‚Üê storing the count
    print(f"{table}: {count_df.iloc[0, 0]}")
```

```python
table_join = []  # use list, not dictionary

for table in tables_df['Tables_in_employee_hiring_db']:
    query = f"""
        SELECT 
            ID, FP, `BU Region`, Approved, `On hold`, `Sourcing start`,
            `Interview start`, `Interview end`, Offered, Filled, Status
        FROM {table}
    """
    employee_hiring_summary = pd.read_sql(query, conn)
    table_join.append(employee_hiring_summary)

# Combine all into one dataframe
final_df = pd.concat(table_join, ignore_index=True)
```
